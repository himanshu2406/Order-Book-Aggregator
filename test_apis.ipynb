Debugging notebook to understand responses from both APIs

```python
import requests
import json
```

```python
coinbase_pro_url = 'https://api.exchange.coinbase.com/products/BTC-USD/book?level=2'
coinbase_resp = requests.get(coinbase_pro_url)
coinbase_resp_payload = json.loads(coinbase_resp.text)
coinbase_resp_payload
```

```python
cb_bids = coinbase_resp_payload.get('bids')
cb_bids[:10]
# bids/asks arrays seems to have - price, quantity, unknown var format
```

```python
cb_asks = coinbase_resp_payload.get('asks')
cb_asks[:10]
```

```python
snapshot_time = coinbase_resp_payload.get('time')
snapshot_time
```
